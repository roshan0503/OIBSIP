# Data Loading and Cleaning

```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv('menu.csv')
```

```python
# Inspect the data
print(df.head())
print(df.info())
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values:\n", missing_values)
```

# Descriptive Statistics

```python
# Calculate basic statistics for numerical columns
desc_stats = df.describe()
print(desc_stats)
```

# Visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Distribution of categories
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Category')
plt.title('Distribution of Food Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Histogram for Calories
plt.figure(figsize=(10, 6))
sns.histplot(df['Calories'], bins=20, kde=True)
plt.title('Distribution of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot for Calories by Category
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='Category', y='Calories')
plt.title('Calories Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Calories')
plt.xticks(rotation=90)
plt.show()
```

```python
# Heatmap for correlations
plt.figure(figsize=(15, 10))
numeric_df = df.select_dtypes(include=['float64', 'int64'])  # Select only numeric columns
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# Product Analysis

```python
# Identify the most and least caloric items
most_caloric_items = df[['Item', 'Calories']].sort_values(by='Calories', ascending=False).head(10)
least_caloric_items = df[['Item', 'Calories']].sort_values(by='Calories').head(10)
```

```python
# Identify items with the highest and lowest fat content
highest_fat_items = df[['Item', 'Total Fat']].sort_values(by='Total Fat', ascending=False).head(10)
lowest_fat_items = df[['Item', 'Total Fat']].sort_values(by='Total Fat').head(10)
```

```python
# Identify items with the highest sodium content
highest_sodium_items = df[['Item', 'Sodium']].sort_values(by='Sodium', ascending=False).head(10)
```

```python
# Visualizations
plt.figure(figsize=(10, 6))
sns.barplot(data=most_caloric_items, x='Calories', y='Item')
plt.title('Top 10 Most Caloric Items')
plt.xlabel('Calories')
plt.ylabel('Item')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=least_caloric_items, x='Calories', y='Item')
plt.title('Top 10 Least Caloric Items')
plt.xlabel('Calories')
plt.ylabel('Item')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=highest_fat_items, x='Total Fat', y='Item')
plt.title('Top 10 Items with Highest Fat Content')
plt.xlabel('Total Fat (g)')
plt.ylabel('Item')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=lowest_fat_items, x='Total Fat', y='Item')
plt.title('Top 10 Items with Lowest Fat Content')
plt.xlabel('Total Fat (g)')
plt.ylabel('Item')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=highest_sodium_items, x='Sodium', y='Item')
plt.title('Top 10 Items with Highest Sodium Content')
plt.xlabel('Sodium (mg)')
plt.ylabel('Item')
plt.show()
```

High-Calorie and High-Fat Items:

The most caloric and high-fat items are predominantly breakfast items and large servings of McNuggets. These items can be highlighted to customers who are looking for high-energy meals.
Reducing portion sizes or offering low-calorie alternatives can help attract health-conscious customers.

Low-Calorie and Low-Fat Items:

Drinks like Diet Dr Pepper, various sizes of Iced Tea, and Dasani Water Bottle are great options for customers looking to cut down on calories.
Introducing more low-fat items in the main course categories can cater to health-conscious customers.

High Sodium Content:

Many of the high-sodium items overlap with high-calorie and high-fat items, particularly breakfast meals and McNuggets.
Reducing the sodium content in these items could improve their nutritional profile and appeal to a wider customer base concerned about their salt intake.

Visual Presentation:

Prominently feature nutritional information on menus to help customers make informed choices.
Use visuals like the bar charts created above to highlight healthy options in marketing materials.

By focusing on these recommendations, the retail business can attract a more diverse customer base while promoting healthier eating habits.
